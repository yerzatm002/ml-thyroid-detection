## 1-кезең — Мәліметтерді талдау және алдын ала өңдеу

Бұл кезеңде біз пациенттердің гормондық және клиникалық мәліметтеріндегі артық/жетіспейтін немесе аномальды мәндерді тексеріп, оларды модельге жарамды форматқа келтіреміз.  
Категориялық сипаттамалар кодталады, ал мақсатты (`class`) баған машиналық оқыту үшін бинарлық форматқа түрлендіріледі.  
Нәтижесінде `X` (түсіндірмелі белгілер) және `y` (мақсатты белгі) дайын болады.

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder

# Көшірме жасау (оригиналды бұзбау үшін)
df_clean = df.copy()

# 1. class бағанын тексеру
df_clean = df_clean[df_clean['class'].notnull()]
target_encoder = LabelEncoder()
df_clean['class_encoded'] = target_encoder.fit_transform(df_clean['class'])

# 2. Таза сандық бағандар (TSH, T3, т.б.) — тек оң сандар
hormone_cols = ['TSH', 'T3', 'TT4', 'T4U', 'FTI']
for col in hormone_cols:
    df_clean = df_clean[df_clean[col].isnull() | (df_clean[col] >= 0)]

# 3. Категориялық бағандарды анықтау
binary_cols = [
    'on thyroxine', 'query on thyroxine', 'on antithyroid medication',
    'sick', 'pregnant', 'thyroid_surgery', 'I131_treatment', 'lithium',
    'goitre', 'tumor', 'hypopituitary', 'psych'
]
categorical_cols = ['sex', 'referral_source']

# 4. Категориялық бағандарды кодтау
le = LabelEncoder()
for col in binary_cols + categorical_cols:
    df_clean[col] = le.fit_transform(df_clean[col].astype(str))

# 5. Қажетсіз бағандарды алып тастау
drop_cols = ['TBG_measured', 'dataset_name', 'patient_id', 'class']
df_clean.drop(columns=drop_cols, inplace=True, errors='ignore')

# 6. Финалдық мәліметтер X және y
X = df_clean.drop(columns=['class_encoded'])
y = df_clean['class_encoded']

# 7. Тексеру нәтижесі
print(f"Қалыптасқан деректер: {X.shape[0]} үлгі, {X.shape[1]} белгі")
print(f"Мақсатты класстар: {y.unique()} → {target_encoder.classes_}")
```

## 2.1 — Гормондар үлестірімін талдау

Бұл бөлімде негізгі гормондардың (`TSH`, `T3`, `TT4`, `FTI`) үлестірімдерін визуализациялаймыз.  
Мақсат — қалыпты және атипиялық мәндерді анықтау, сонымен қатар келесі модельдеуге дайындық жасау.  
TSH логарифмдік шкалада көрсетіледі, себебі ол жиі лог-нормалды таралған. Қалған гормондар — KDE қисықтары арқылы көрінеді.

```python
import matplotlib.pyplot as plt
import seaborn as sns

hormones_to_plot = ['TSH', 'T3', 'TT4', 'FTI']
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
axes = axes.flatten()

for i, hormone in enumerate(hormones_to_plot):
    ax = axes[i]
    data = df_clean[hormone].dropna()

    if hormone == 'TSH':
        data = np.log1p(data)  # log(1 + x) to handle 0s
        sns.histplot(data, kde=True, bins=50, ax=ax, color='skyblue')
        ax.set_title("TSH (лог шкала)")
        ax.set_xlabel("log(TSH + 1)")
    else:
        sns.histplot(data, kde=True, bins=40, ax=ax, color='salmon')
        ax.set_title(f"{hormone} үлестірімі")
        ax.set_xlabel(hormone)

    ax.set_ylabel("Пациент саны")

plt.suptitle("Негізгі гормондардың үлестірімі", fontsize=16)
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

## 2.2 — Гормондар арасындағы корреляциялық талдау

Бұл бөлімде біз TSH, T3, TT4, T4U, FTI сияқты негізгі гормондардың арасындағы өзара байланыстарды анықтаймыз.

Бұл әдіс арқылы алдын-ала ең ықтималды маңызды белгілерді таңдай аламыз.

График 10 — Бастапқы белгілер маңыздылығы (корреляция бойынша)

```python
# Тек сандық бағандар
numeric_df = df_clean.select_dtypes(include=['number']).copy()

# 'class_encoded'-тен басқа бағандар (X белгілері)
feature_cols = numeric_df.columns.drop('class_encoded')

# Корреляцияны есептеу
correlations = numeric_df[feature_cols].corrwith(numeric_df['class_encoded']).abs()

# DataFrame түрінде
corr_df = correlations.sort_values(ascending=True).reset_index()
corr_df.columns = ['feature', 'correlation']

# График
plt.figure(figsize=(10, 8))
sns.barplot(
    data=corr_df,
    x='correlation',
    y='feature',
    palette='magma'
)
plt.title("Бастапқы белгілердің маңыздылығы (class-пен корреляция негізінде)", fontsize=14)
plt.xlabel("Абсолюттік корреляция")
plt.ylabel("Белгілер")
plt.tight_layout()
plt.show()
```

## 2.6 — Радиациялық диаграмма (Spider Chart)

Бұл бөлімде әртүрлі диагноз классындағы пациенттердің гормондық профилін салыстырамыз.  
Радиациялық диаграмма (немесе паук-сұлба) арқылы әр гормонның орташа мәні бір көзқараста салыстырылады.  
Мақсат — әрбір диагноз класына тән гормон үлгісін визуалды түрде көрсету.

График 11 — Диагноз кластарындағы гормондық профиль айырмашылықтары

```python
import matplotlib.pyplot as plt
import numpy as np

# 1. Таңдалатын гормондық белгілер
hormones = ['TSH', 'T3', 'TT4', 'T4U', 'FTI']
class_names = target_encoder.classes_[:4]  # Шектеу: 4 класс ғана графикке түседі

# 2. Орташа мәндерді есептеу (тек таңдаулы класстар үшін)
mean_profiles = {}
for class_name in class_names:
    class_code = target_encoder.transform([class_name])[0]
    profile = df_clean[df_clean['class_encoded'] == class_code][hormones].mean()
    mean_profiles[class_name] = profile

# 3. Қиылыстыру үшін белгі координаталары
categories = list(hormones)
N = len(categories)
angles = np.linspace(0, 2 * np.pi, N, endpoint=False).tolist()
angles += angles[:1]  # тұйықтау үшін

# 4. График
plt.figure(figsize=(8, 8))
ax = plt.subplot(111, polar=True)

# Әр класс үшін сызық сызу
for class_name, values in mean_profiles.items():
    data = values.tolist()
    data += data[:1]  # тұйықтау
    ax.plot(angles, data, label=class_name)
    ax.fill(angles, data, alpha=0.1)

# Ось атаулары
ax.set_xticks(angles[:-1])
ax.set_xticklabels(categories, fontsize=12)

# Радиус
ax.set_title("Гормон профилі — диагноз класы бойынша", fontsize=14, pad=20)
ax.legend(loc='upper right', bbox_to_anchor=(1.2, 1.1))
plt.tight_layout()
plt.show()
```

## 2.8 — t-SNE визуализациясы (кластардың тығыз орналасуы)

`t-SNE` — жоғары өлшемді мәліметтерді төмен өлшемге (2D немесе 3D) қысқартатын алгоритм.  
Бұл әдіс деректер ішіндегі жасырын құрылымдарды визуалды түрде бөліп көрсету үшін қолданылады.  
`t-SNE` көмегімен біз кластардың (диагноздардың) бір-бірінен бөлек немесе ішкі топтарға бөліну мүмкіндігін көреміз.

График 13 — t-SNE негізіндегі визуализация (`class_encoded` бойынша)

```python
from sklearn.manifold import TSNE
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# ↓↓↓ 1. Подвыборка ↓↓↓
X_sample = X.sample(n=10000, random_state=42)
y_sample = y.loc[X_sample.index]

# ↓↓↓ 2. Импутация и масштаб ↓↓↓
imputer = SimpleImputer(strategy='mean')
X_imputed = imputer.fit_transform(X_sample)

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_imputed)

# ↓↓↓ PCA ↓↓↓
pca = PCA(n_components=10, random_state=42)
X_pca = pca.fit_transform(X_scaled)

# ↓↓↓ t-SNE ↓↓↓
tsne = TSNE(n_components=2, perplexity=30, learning_rate=200, n_iter=500, random_state=42)
X_tsne = tsne.fit_transform(X_pca)

# ↓↓↓ 5. Визуализация ↓↓↓
tsne_df = pd.DataFrame(X_tsne, columns=['Dim1', 'Dim2'])
tsne_df['class'] = y_sample.map(dict(enumerate(target_encoder.classes_)))

plt.figure(figsize=(10, 7))
sns.scatterplot(
    data=tsne_df, x='Dim1', y='Dim2',
    hue='class', palette='tab10', alpha=0.8, s=50
)
plt.title("t-SNE визуализациясы (үлгі бойынша, 10K)", fontsize=14)
plt.xlabel("Өлшем 1")
plt.ylabel("Өлшем 2")
plt.legend(loc='best', bbox_to_anchor=(1.05, 1))
plt.tight_layout()
plt.show()
```

## 3.0 — Финалдық деректерді тазарту және дайындау (Model-ready)

Бұл бөлімде біз барлық деректерді біріктіріп, модельдерге жарамды форматқа келтіреміз.  
Мақсат:  
- Барлық категориалды айнымалылар → сандық  
- NaN мәндер → толтырылу  
- Қажетсіз бағандар → жою  
- Масштабтау → барлық белгілерді бір диапазонға түсіру  
Нәтижесінде `X` және `y` толығымен дайын болады.

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.impute import SimpleImputer

# 1. Көшірме жасау
df_model = df_clean.copy()

# 2. Мақсатты айнымалыны дайындау
y = df_model['class_encoded']

# 3. Қажет емес бағандарды жою (бұрын қолданған: class, dataset_name, т.б.)
drop_cols = ['class', 'dataset_name', 'patient_id', 'class_encoded', 'age_group']
df_model.drop(columns=[col for col in drop_cols if col in df_model.columns], inplace=True)

# 4. Категориялық бағандарды анықтау
cat_cols = df_model.select_dtypes(include='object').columns.tolist()
bin_cols = [
    'on thyroxine', 'query on thyroxine', 'on antithyroid medication', 'sick', 'pregnant',
    'thyroid_surgery', 'I131_treatment', 'lithium', 'goitre', 'tumor', 'hypopituitary', 'psych'
]

# 5. Барлық категориалды бағандарды LabelEncoder арқылы сандыққа айналдыру
label_enc = LabelEncoder()
for col in cat_cols + bin_cols:
    if col in df_model.columns:
        df_model[col] = label_enc.fit_transform(df_model[col].astype(str))

# 6. NaN → орташа мән арқылы толтыру
imputer = SimpleImputer(strategy='mean')
X_imputed = imputer.fit_transform(df_model)

# 7. Масштабтау (0-mean, unit-variance)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_imputed)

# 8. Финалдық: X, y дайын
X = pd.DataFrame(X_scaled, columns=df_model.columns)

print(f"X пішіні: {X.shape}, y пішіні: {y.shape}")
```

---
 Бұл бөлімнің нәтижесі бойынша біз модельдердің салыстырмалы тиімділігін бағалап, ең үздігін таңдаймыз.


```python
from sklearn.linear_model import LogisticRegression

# Модель құру
log_reg = LogisticRegression(
    max_iter=1000,                  # Итерация саны (үлкен dataset үшін маңызды)
    class_weight='balanced',        # Класстар арасындағы теңсіздікке бейімделу
    solver='lbfgs',                 # Көпклассты жағдай үшін тұрақты solver
    multi_class='multinomial',      # Softmax әдісі (бір-біріне ұқсамайтын класстар үшін)
    random_state=42
)

# Модельді оқыту
log_reg.fit(X_train, y_train)

# Модель дайын
print(" LogisticRegression моделі оқытылды.")
```

```python
from sklearn.ensemble import RandomForestClassifier

# Модель құру
rf = RandomForestClassifier(
    n_estimators=100,           # Ағаш саны (көбейген сайын тұрақтылық артады)
    max_depth=None,             # Барлық тереңдікке дейін үйренеді
    class_weight='balanced',    # Класстар дисбалансын ескеріп оқыту
    random_state=42             # Қайталанғыштық үшін
)

# Оқыту
rf.fit(X_train, y_train)

# Модель дайын
print(" RandomForestClassifier моделі оқытылды.")
```

```python
from xgboost import XGBClassifier

# XGBoost моделі
xgb = XGBClassifier(
    n_estimators=100,            # Бустинг итерация саны
    learning_rate=0.1,           # Оқу қадамы (0.05–0.3 арасында жақсы жұмыс істейді)
    max_depth=6,                 # Әр ағаштың максималды тереңдігі
    objective='multi:softprob',  # Мультиклассты softmax классификация
    num_class=len(set(y_train)), # Класстар саны
    eval_metric='mlogloss',      # Логарифмдік шығын функциясы (жақсырақ convergence)
    use_label_encoder=False,     # Ескі ескерту хабарламаларын болдырмау
    random_state=42              # Қайталанымдылық
)

# Оқыту
xgb.fit(X_train, y_train)

# Дайын
print(" XGBoostClassifier моделі оқытылды.")
```

## 5.3 — ROC-кривые (мультиклассты)

Бұл бөлімде біз ROC-кривойларды салыстырамыз. Бұл модельдің кластарды ажырата білу қабілетін көрсетеді.

```python
from sklearn.metrics import roc_curve, auc
from sklearn.preprocessing import label_binarize
from itertools import cycle
import matplotlib.pyplot as plt
import numpy as np

# 1. Binarize класстар (нүктелі түрде)
classes = np.unique(y)
y_test_bin = label_binarize(y_test, classes=classes)
n_classes = y_test_bin.shape[1]

# 2. Әр модельге бір-бірден tpr, fpr есептеу
plt.figure(figsize=(10, 8))

colors = cycle(['darkorange', 'green', 'blue'])

models = {
    "Logistic Regression": log_reg,
    "Random Forest": rf,
    "XGBoost": xgb
}


for (name, model), color in zip(models.items(), colors):
    y_score = model.predict_proba(X_test)

    # Мультиклассты ROC-AUC
    fpr = dict()
    tpr = dict()
    roc_auc = dict()
    
    for i in range(n_classes):
        fpr[i], tpr[i], _ = roc_curve(y_test_bin[:, i], y_score[:, i])
        roc_auc[i] = auc(fpr[i], tpr[i])

    # Macro-average ROC
    all_fpr = np.unique(np.concatenate([fpr[i] for i in range(n_classes)]))
    mean_tpr = np.zeros_like(all_fpr)

    for i in range(n_classes):
        mean_tpr += np.interp(all_fpr, fpr[i], tpr[i])
    mean_tpr /= n_classes

    macro_auc = auc(all_fpr, mean_tpr)
    plt.plot(all_fpr, mean_tpr, color=color, label=f"{name} (AUC = {macro_auc:.4f})", linewidth=2)

# 3. Безендіру
plt.plot([0, 1], [0, 1], 'k--', linewidth=1)
plt.title(" ROC-кривые (мультиклассты)", fontsize=15)
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.legend(loc='lower right')
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

## 5.4 SHAP анализі — Модель шешімдерін интерпретациялау

 **SHAP (SHapley Additive exPlanations)** — бұл әрбір feature-дің (белгінің) модель шешіміне әсерін санап, көрнекі түрде түсіндіретін қуатты құрал.

Мақсаты — модельдің **"неге осылай шешім қабылдағанын"** түсіну.

---

###  SHAP summary plot — feature importance (орташа әсер)
- Әрбір белгі үшін орташа SHAP мәні есептеледі
- Қай feature модельге қаншалықты әсер ететінін көрсетеді
- 🟦 Теріс әсер (ауру болу ықтималдығы төмендетеді)
- 🟥 Оң әсер (ауру болу ықтималдығын арттырады)

---

###  SHAP distribution plot — барлық пациенттерге әсері
- Әрбір нүкте — бір пациент
- Горизонталь ось — feature-дің әсері (SHAP value)
- Түстер — белгінің мәні (қызыл = жоғары, көк = төмен)

Бұл график модельдің шешімдері қандай факторларға сүйенетінін **жеке деңгейде** көруге мүмкіндік береді.

---

###  Неліктен SHAP қолданамыз?
- Модельді қара жәшік күйінен шығарып, **түсінікті** қылады
- Врачу/аналитику шешімді **түсіндіруге** мүмкіндік береді
- Деректер ғылымында интерпретация — маңызды қадам

---

```python
import shap

# SHAP кітапханасын инициализациялау
explainer = shap.Explainer(xgb)
shap_values = explainer(X_train)

# SHAP importance summary plot (feature importance)
shap.summary_plot(shap_values, X_train, plot_type="bar")
```
